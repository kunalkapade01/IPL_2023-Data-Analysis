```python
#Import all libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#Read the csv file:
df = pd.read_csv("D:\Dataset files_ML\IPL 2023/deliveries.csv")
```

```python
df["season"].value_counts()
```

```python
df.drop(columns=["season","penalty","other_wicket_type","other_player_dismissed"], inplace=True)
```

```python
df["total_runs"] = df["runs_off_bat"]+df["extras"]
```

```python
df.head()
```

```python
bowler_wicket = df["wicket_type"].value_counts().index
bowler_wicket = ['caught', 'bowled', 'lbw','caught and bowled', 'stumped', 'hit wicket']
```

```python
df["bowler_wicket"] = list(map(lambda x: x if x in bowler_wicket else np.nan, df["wicket_type"]))
```

```python
df.head()
```

```python
df["bowler_wicket"].value_counts()
```

```python
df["wicket_type"].value_counts()
```

```python
df.head()
```

```python
#To find the orange-cap player Name:
orange_cap = df.groupby("striker")["runs_off_bat"].sum()
orange_cap = orange_cap.sort_values(ascending=False)
orange_cap = orange_cap.index[0]
orange_cap
```

```python
#To find the purple-cap player name:
purple_cap = pd.crosstab(df["bowler"], df["bowler_wicket"]).T
purple_cap = purple_cap.sum().sort_values(ascending=False).index[0]
purple_cap
```

```python
#To find the Highest score:(HS)
HS_in_one_match = df.groupby(["match_id","striker"])[["runs_off_bat"]].sum()
HS_in_one_match = HS_in_one_match.sort_values(by=["runs_off_bat"], ascending=False)
HS_in_one_match.reset_index(inplace=True)
HS_in_one_match.head()
```

```python
wickets = ["caught", "bowled", "lbw", "caught and bowled", "stumped", "hit wicket"]
```

```python
df["total_bowler_wicket"] = list(map(lambda x: 1 if x in  wickets else np.nan, df["wicket_type"]))
```

```python
#To find Best bowling figure:
best_bowling_figures = df.groupby(["match_id","bowler"])[["runs_off_bat","total_bowler_wicket"]].sum()
best_bowling_figures.reset_index(inplace=True)
best_bowling_figures.sort_values(["total_bowler_wicket","runs_off_bat"], ascending=[False, True], inplace=True)
best_bowling_figures.head()
```

```python
#To find the Average of best batting player:
best_batting_avg = df.groupby(["match_id","striker"])[["runs_off_bat"]].sum()
best_batting_avg.reset_index(inplace=True)
best_batting_avg.groupby("striker")["runs_off_bat"].mean().sort_values(ascending=False).head()
#best_batting_avg.head()
```

```python
df[df["striker"]=="RK Singh"]["runs_off_bat"].sum() / len(df[df["striker"]=="RK Singh"]["match_id"].unique())
```

```python
#Find the Average of best bowling player:
best_bowling_avg = df.groupby(["match_id","bowler"])[["runs_off_bat", "total_bowler_wicket"]].sum().reset_index()
best_bowling_avg = best_bowling_avg.groupby("bowler")["runs_off_bat"].sum() / best_bowling_avg.groupby("bowler")["total_bowler_wicket"].sum()
best_bowling_avg.sort_values(ascending=True).head()
```

```python
#Find most Hundred(century(100s)) Runs in IPL 2023:
century = df.groupby(["match_id","striker"])[["runs_off_bat"]].sum().reset_index()
most_century = century[century["runs_off_bat"] >= 100]["striker"].value_counts()
most_century = pd.DataFrame(data=most_century)
most_century
```

```python
#Find most 50s(Half century) Runs:
half_century = df.groupby(["match_id","striker"])[["runs_off_bat"]].sum().reset_index()
most_fifties = half_century[(half_century["runs_off_bat"] >= 50) & (half_century["runs_off_bat"] <= 100)]["striker"].value_counts()
most_fifties = pd.DataFrame(data=most_fifties)
most_fifties.head()
```
